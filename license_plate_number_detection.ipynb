```python
import cv2
import numpy as np

#cap = cv2.imread('20180402113123_NumberPlate_Swift.jpg')
#cap = cv2.VideoCapture('videoplay.mp4')
number_plate_classfier=cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')

count=0;
#while(cap.isOpened()==True):
while True:
    frame = cv2.imread('car2.jpg')
    #ret, frame = cap.read()
    #gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    plate = number_plate_classfier.detectMultiScale(gray, 1.3, 5)
 
    for (x,y,w,h) in plate:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),2)
        roi = frame[y:y+h, x:x+w]
        file_name_path = './plate/' + str(count) + '.jpg'
        cv2.imwrite(file_name_path, roi)
        count += 1
    cv2.imshow('img',frame)
    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break
        
#frame.release()
cv2.destroyAllWindows()   
```

```python
import requests
import xmltodict
import json
import re
import easyocr
#import warnings
#warnings.filterwarnings("ignore")
```

```python
def license_img(img):
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    filtering = cv2.bilateralFilter(gray,11,17,17)
    edge_detect = cv2.Canny(filtering,30,100)
    cntrs=cv2.findContours(edge_detect.copy(),cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    contours = imutils.grab_contours(cntrs)
    contours = sorted(contours,key=cv2.contourArea, reverse = True)[:10]
    contour_points = None
    for c in contours:
        approx = cv2.approxPolyDP(c, 10 , True)
       # if our approximated contour has four points, then
       # we can assume that we have found our screen
        if len(approx) == 4:
            contour_points = approx
            break
            
    mask = np.zeros(gray.shape,np.uint8)
    
    new_image = cv2.drawContours(mask,[contour_points],0,255,-1,)
    new_image = cv2.bitwise_and(img,img,mask=mask)
    
    (x, y) = np.where(mask == 255)
    (x1, y1) = (np.min(x), np.min(y))
    (x2, y2) = (np.max(x), np.max(y))
    crop_img = gray[x1:x2+1, y1:y2+1]
    
    read =  easyocr.Reader(['en'])
    result = read.readtext(crop_img)
    
    return result
```

```python
import cv2
import numpy as np
from os import listdir
from os.path import isfile, join

# Get the training data we previously made
data_path = './plate/'
onlyfiles = [f for f in listdir(data_path) if isfile(join(data_path, f))]

# Create arrays for training data and labels
license_plate= []

# Open training images in our datapath
# Create a numpy array for training data
for i, files in enumerate(onlyfiles):
    image_path = data_path + onlyfiles[i]
    images = cv2.imread(image_path)
    
    nplate=license_img(images)
    #print(number_plate)
    if nplate!=None and len(nplate)==1:
        if (nplate[0][-1]>0.3):
            nplate=nplate[0][-2].replace(" ","")
            license_plate.append(nplate)
            #print(license_plate_number)
```
